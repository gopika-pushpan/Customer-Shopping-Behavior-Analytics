```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("customer_data.db")
```

```python
pd.read_sql("""
SELECT SUM("Purchase Amount (USD)") AS Total_Revenue
FROM customers
""", conn)
```

**Insight:** This query shows the total revenue generated from all customer purchases.


```python
pd.read_sql("""
SELECT Season, SUM("Purchase Amount (USD)") AS Revenue
FROM customers
GROUP BY Season
ORDER BY Revenue DESC
""", conn)
```

**Insight:** Identifies which season contributes the highest revenue.


```python
pd.read_sql("""
SELECT Category, SUM("Purchase Amount (USD)") AS Revenue
FROM customers
GROUP BY Category
ORDER BY Revenue DESC
""", conn)
```

```python
pd.read_sql("""
SELECT Gender, SUM("Purchase Amount (USD)") AS Revenue
FROM customers
GROUP BY Gender
""", conn)
```

```python
pd.read_sql("""
SELECT "Subscription Status", SUM("Purchase Amount (USD)") AS Revenue
FROM customers
GROUP BY "Subscription Status"
""", conn)
```

```python
pd.read_sql("""
SELECT "Shipping Type", SUM("Purchase Amount (USD)") AS Revenue
FROM customers
GROUP BY "Shipping Type"
""", conn)
```

```python
pd.read_sql("""
SELECT
CASE
    WHEN Age < 20 THEN 'Teen'
    WHEN Age BETWEEN 20 AND 30 THEN 'Young Adult'
    WHEN Age BETWEEN 31 AND 50 THEN 'Adult'
    ELSE 'Senior'
END AS Age_Group,
SUM("Purchase Amount (USD)") AS Revenue
FROM customers
GROUP BY Age_Group
ORDER BY Revenue DESC
""", conn)
```

## Key Business Insights

- Adult customers (3150) contribute the highest revenue.
- Certain seasons outperform others in total sales.
- Subscription users show higher overall spending.
- Standard shipping is the most preferred option.
- Gender-wise spending patterns show noticeable differences.
